## Cost Functions and Gradient Descent

One of the fundamental concepts in machine learning is the cost function and the optimization technique known as gradient descent. These concepts are crucial for training models and improving their performance.

### The Need for Cost Functions

A cost function, also known as a loss function or objective function, quantifies how well a machine learning model is performing. It measures the difference between the predicted values (output) and the actual values (ground truth) for a given set of input data. The goal is to minimize this difference, indicating that the model is making accurate predictions.

### The Anatomy of a Cost Function

Mathematically, a cost function can be represented as **C()**, where  represents the model's parameters (weights and biases). The choice of the cost function depends on the specific problem you are trying to solve. Common cost functions include Mean Squared Error (MSE) for regression problems and Cross-Entropy for classification problems.

### Mean Squared Error (MSE)

MSE is commonly used for regression tasks, where you predict continuous values. It calculates the average of the squared differences between the predicted values and the actual values.

## Gradient Descent: The Optimization Technique

Now that we have a measure of how well our model is performing through the cost function, the next step is to optimize our model to minimize this cost. This is where gradient descent comes into play.

Gradient descent is an iterative optimization algorithm that adjusts the model's parameters () to minimize the cost function (C()). It does this by calculating the gradient of the cost function with respect to the model's parameters and updating the parameters in the opposite direction of the gradient.

Here are the key steps of gradient descent:

1. **Initialization**: Start with initial values for the model's parameters ().

2. **Calculate the Gradient**: Compute the gradient of the cost function with respect to . This gradient points in the direction of the steepest increase in the cost function.

3. **Update Parameters**: Adjust the parameters by moving in the opposite direction of the gradient. The size of this step is controlled by a hyperparameter called the learning rate ().

4. **Repeat**: Continue steps 2 and 3 for a specified number of iterations or until the cost converges to a minimum value.

The learning rate () is a critical hyperparameter that determines the step size in each iteration. If it's too large, the algorithm may overshoot the minimum, and if it's too small, it may converge very slowly or get stuck in local minima.

```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
y = np.array([4,7,6,8,10,12,16])
X = np.array([6,7,8,9,10,11,11])
```

```python
m = 0 # gradiant
c = 0 # intercept
alpha = 0.1 # learning rate
iteration = 1000
n = len(X) # No. of data points
```

```python
for i in range(iteration):
    # Cost function
    mse = (1/n)*sum(((m * X + c) - y)**2)

    # Update params
    m = m - alpha*((2/n)*sum((m * X + c)-y))
    c = c - alpha*((2/n)*sum((m * X + c)-y))
    print(f"error = {mse} ,iteration = {i}")

print(f"\nm = {m} \nc = {c}")
```
