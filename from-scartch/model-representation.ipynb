```python
import numpy as np
import numpy.typing
import matplotlib.pyplot as plt
```

```python
x_train = np.array([1.0, 2.0, 3.0])
y_train = np.array([300.0, 500.0, 700.0])

print(f'x_train = { x_train }')
print(f'y_train = { y_train }')
```

## Number of training examples  m

```python
print(f'x_train.shape: { x_train.shape }')
m = x_train.shape[0] # == len(x_train)
print(f'Number of training examples is: { m }')
```

```python
i: int = 0;

for i in range(m):
  x_i = x_train[i]
  y_i = x_train[i]
  print(f'(x^({ i }), y^({ i })) = ({ x_i }, { y_i })')
```

## Plotting the data
- use `scatter()` in `matplotlib`

```python
plt.scatter(x_train, y_train, marker='x', c='r')
plt.title("Housing Prices")
plt.ylabel('Price (in 1000s of dollars)')
plt.xlabel('Size (1000 sqft)')
plt.show()
```

## Model Function

$$ f_{w,b}(x^{(i)}) = wx^{(i)} + b \tag{1}$$


```python
w = 200
b = 100
print(f"w: {w}")
print(f"b: {b}")
```

for $x^{(0)}$, `f_wb = w * x[0] + b`

for $x^{(1)}$, `f_wb = w * x[1] + b`

for $x^{(2)}$, `f_wb = w * x[2] + b`

```python
def compute_model_output(x: numpy.ndarray, w: int , b: int) -> np.array:
  '''
  Compute the prediction of a linear model

  Args:
    x (ndarray (m,)): Data, m examples
    w, b (scalar)   : model parameters

  Returns
    f_wb (ndarray (m,)): model predictions
  '''

  m = x.shape[0]
  f_wb = np.zeros(m)
  for i in range(m):
    f_wb[i] = w * x[i] + b
  
  return f_wb
```

```python
tmp_f_wb = compute_model_output(x_train, w, b)
tmp_f_wb
```

#### Plot the prediction

```python
plt.plot(x_train, tmp_f_wb, c='b', label='Our Predictions')
plt.scatter(x_train, y_train, marker='x', c='r', label='Actual Values')

plt.title('Housing Prices')
plt.ylabel('Price (in 1000s of dollars)')
plt.xlabel('Size (1000 sqft)')
plt.legend()
plt.show()
```

## Prediction

predict the price of a house with 1200 sqft. Since the units of  $x$ are in 1000's of sqft,  $x$ is 1.2.
